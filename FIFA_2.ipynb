```python
import pandas as pd 
import numpy as np
```

# - Estandariazacion de nombres en mis columnas

```python

df.columns=df.columns.str.lower().str.replace(' ','_')
```

```python
##quite las columnas que no iba a ocupar o me parecian redundante
```

```python
df.drop(['w/f', 'ir','sm','a/w','d/w','lcb','pac', 'sho', 'pas', 'cb', 'dri', 'def' , 'phy', 'ls', 'rcb','st' ,'rs', 'lw','lf','cf','rb', 'rf' , 'rw','lam', 'cam','ram' , 'gk', 'lm' , 'lcm','cm' ,'rcm', 'rm', 'lwb', 'cdm','rdm', 'rwb', 'lb' ,'gender' ], axis=1, inplace=True)
```

```python
## quite el valor de euros .
```

```python
df['value']=df['value'].str.replace('€','')
```

```python
df['wage']=df['wage'].str.replace('€','')
```

```python
df['release_clause']=df['release_clause'].str.replace('€','')
```

```python
## quite los na
```

```python
df.dropna()
```

```python
df.dtypes
```

```python
df_fifa = df.dropna()
df_fifa.head(8)
```

```python
##cambie el tipo de dato que tenia. CAMBIE A ENTERO
```

```python
df_fifa = df_fifa.astype({'volleys': 'int64' , 'curve': 'int64', 'balance': 'int64', 'jumping': 'int64' , 'interceptions': 'int64' , 'positioning': 'int64' , 'vision': 'int64' , 'sliding_tackle': 'int64'})
df_fifa.dtypes
```

```python
df_fifa.info()
```
